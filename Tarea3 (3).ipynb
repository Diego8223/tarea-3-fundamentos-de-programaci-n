Para este ejercicio se deben almacenar los factores de conversión en un vector (tipo de dato ```list``` de Python) para luego:

- Solicitar al usuario que escoja el tipo de conversión que quiere realizar. **Para esta nueva versión de algoritmo se sugiere hacer esta solicitud únicamente con ``input`` y no con formularios de Colab**

````python
#Desarrolle su ejercicio aquí.

# Lista de factores de conversión
factores_conversion = [
    1000,        # 0: Kilómetro a metro
    0.01,        # 1: Centímetro a metro
    0.62,        # 2: Kilómetro por hora a millas por hora
    20,          # 3: Nota en Educatic a nota en Canvas
    0.05,        # 4: Nota en Canvas a nota en Educatic
    1,           # 5: Mililitro a centímetro cúbico
    "Actualizar",# 6: Dólar (USA) a peso colombiano
    "Actualizar",# 7: Peso colombiano a dólar (USA)
    0.45,        # 8: Libra a kilogramo
    17987539.67  # 9: Minuto luz a kilómetro
]

# Función para realizar conversiones
def realizar_conversion():
    while True:
        print("\nSeleccione el tipo de conversión:")
        print("""
        0: Kilómetro a metro
        1: Centímetro a metro
        2: Kilómetro por hora a millas por hora
        3: Nota en Educatic a nota en Canvas
        4: Nota en Canvas a nota en Educatic
        5: Mililitro a centímetro cúbico
        6: Dólar (USA) a peso colombiano (actualizar manualmente)
        7: Peso colombiano a dólar (USA) (actualizar manualmente)
        8: Libra a kilogramo
        9: Minuto luz a kilómetro
        """)

        # Solicitar la conversión deseada
        try:
            tipo_conversion = int(input("Ingrese el índice del tipo de conversión (0-9): "))
            if tipo_conversion not in range(10):
                print("Por favor, elija un índice válido.")
                continue
        except ValueError:
            print("Debe ingresar un número entero. Inténtelo de nuevo.")
            continue

        # Si es una conversión monetaria, pedir actualización del factor
        if tipo_conversion in [6, 7]:
            try:
                factor_actualizado = float(input(f"Ingrese el factor de conversión para {('Dólar a peso colombiano' if tipo_conversion == 6 else 'Peso colombiano a dólar')}: "))
                factores_conversion[tipo_conversion] = factor_actualizado
            except ValueError:
                print("Debe ingresar un valor numérico válido para el factor de conversión.")
                continue

        # Solicitar el valor a convertir
        try:
            valor = float(input("Ingrese el valor a convertir: "))
        except ValueError:
            print("Debe ingresar un valor numérico. Inténtelo de nuevo.")
            continue

        # Realizar la conversión
        factor = factores_conversion[tipo_conversion]
        resultado = valor * factor
        print(f"El resultado de la conversión es: {resultado}")

        # Preguntar si desea continuar
        continuar = input("¿Desea realizar otra conversión? (s/n): ").strip().lower()
        if continuar != 's':
            print("¡Gracias por usar el programa de conversiones!")
            break

# Ejecutar el programa
realizar_conversion()



````

# (25%) Ejercicio 2
**Vamos a mejorar la solución del Ejercicio 2 - Tarea 2.**

Estudie cómo realizar conversiones entre los sistemas de numeración Hexadecimal, Binario y Decimal. Se sugiere leer el documento *Sistemas de numeración* disponible en el Material complementario del curso.

Programe un algoritmo que solicite al usuario la conversión que desea realizar, que puede ser:
* de Binario a Decimal
* de Hexadecimal a Decimal
* de Decimal a Binario
* de Decimal a Hexadecimal
* de Binario a Hexadecimal
* de Hexadecimal a Binario

Solicite también la cantidad a convertir.

Finalmente, con estructuras de selección y **estructuras repetitivas** aplique el algoritmo necesario para la conversión deseada y presente al usuario la respuesta.

**Nota:** Con el objetivo de practicar los ciclos, no está permitido usar funciones de conversión nativas como ``bin()``, ``hex()``, ``format()``, etc, o funciones similares de librerías.

````python
#Desarrolle su ejercicio aquí.

# Diccionarios para mapear valores de Binario a Decimal y Hexadecimal
hex_digits = "0123456789ABCDEF"
binary_digits = {
    0: "0000", 1: "0001", 2: "0010", 3: "0011", 4: "0100",
    5: "0101", 6: "0110", 7: "0111", 8: "1000", 9: "1001",
    10: "1010", 11: "1011", 12: "1100", 13: "1101", 14: "1110", 15: "1111"
}

# Función para convertir de Binario a Decimal
def bin_to_decimal(bin_num):
    decimal = 0
    power = 0
    for digit in bin_num[::-1]:  # Recorremos el binario de derecha a izquierda
        if digit == '1':
            decimal += 2 ** power
        power += 1
    return decimal

# Función para convertir de Hexadecimal a Decimal
def hex_to_decimal(hex_num):
    decimal = 0
    power = 0
    for digit in hex_num[::-1]:  # Recorremos el hexadecimal de derecha a izquierda
        decimal += hex_digits.index(digit.upper()) * (16 ** power)
        power += 1
    return decimal

# Función para convertir de Decimal a Binario
def decimal_to_bin(decimal):
    bin_num = ""
    while decimal > 0:
        bin_num = str(decimal % 2) + bin_num
        decimal = decimal // 2
    return bin_num if bin_num else "0"

# Función para convertir de Decimal a Hexadecimal
def decimal_to_hex(decimal):
    hex_num = ""
    while decimal > 0:
        hex_num = hex_digits[decimal % 16] + hex_num
        decimal = decimal // 16
    return hex_num if hex_num else "0"

# Función para convertir de Binario a Hexadecimal
def bin_to_hex(bin_num):
    decimal = bin_to_decimal(bin_num)
    return decimal_to_hex(decimal)

# Función para convertir de Hexadecimal a Binario
def hex_to_bin(hex_num):
    decimal = hex_to_decimal(hex_num)
    return decimal_to_bin(decimal)

# Menú de selección
def menu_conversion():
    print("\nSeleccione la conversión que desea realizar:")
    print("1. Binario a Decimal")
    print("2. Hexadecimal a Decimal")
    print("3. Decimal a Binario")
    print("4. Decimal a Hexadecimal")
    print("5. Binario a Hexadecimal")
    print("6. Hexadecimal a Binario")
    print("0. Salir")

# Función principal
def conversor():
    while True:
        menu_conversion()

        try:
            # Solicitar al usuario la opción de conversión
            opcion = int(input("\nIngrese el número de la opción (0 para salir): "))

            if opcion == 0:
                print("Saliendo del programa...")
                break

            if opcion == 1:
                bin_num = input("Ingrese el número binario: ")
                decimal_result = bin_to_decimal(bin_num)
                print(f"El número binario {bin_num} en decimal es {decimal_result}")

            elif opcion == 2:
                hex_num = input("Ingrese el número hexadecimal: ")
                decimal_result = hex_to_decimal(hex_num)
                print(f"El número hexadecimal {hex_num} en decimal es {decimal_result}")

            elif opcion == 3:
                decimal_num = int(input("Ingrese el número decimal: "))
                bin_result = decimal_to_bin(decimal_num)
                print(f"El número decimal {decimal_num} en binario es {bin_result}")

            elif opcion == 4:
                decimal_num = int(input("Ingrese el número decimal: "))
                hex_result = decimal_to_hex(decimal_num)
                print(f"El número decimal {decimal_num} en hexadecimal es {hex_result}")

            elif opcion == 5:
                bin_num = input("Ingrese el número binario: ")
                hex_result = bin_to_hex(bin_num)
                print(f"El número binario {bin_num} en hexadecimal es {hex_result}")

            elif opcion == 6:
                hex_num = input("Ingrese el número hexadecimal: ")
                bin_result = hex_to_bin(hex_num)
                print(f"El número hexadecimal {hex_num} en binario es {bin_result}")

            else:
                print("Opción inválida. Por favor, ingrese un número del 1 al 6.")

        except ValueError:
            print("Error: Ingrese un valor numérico válido.")

# Ejecutar el conversor
conversor()
````

En el [Portal de datos de Colombia](https://www.datos.gov.co/Ambiente-y-Desarrollo-Sostenible/Precipitaciones-Totales-Mensuales/mb4n-6m2g) se publica el volumen de precipitaciones mensuales obtenido en la red de estaciones hidrológicas de la Corporación Autónoma Regional de Cundinamarca - CAR.

En este ejercicio debes realizar un algoritmo que calcule lo siguiente:

1.   El promedio anual de precipitación en el Lago para cada año presentado en la tabla.
2.   El promedio mensual de precipitación en el Lago para cada **mes** presentado en la tabla.
2.   El mes y año que registra menor precipitación.
3.   El año más lluvioso.
4.   El mes más lluvioso.

**Nota:**
1. Los datos presentados en la tabla incluyen 6 años de información recogida en la Estación Meteorológica El Bosque de la cuenca Río Tunjuelo del Municipio de Santafe de Bogotá.

2. Medición de la lluvia: La precipitación o cantidad de lluvia caída se mide ubicando un recipiente de fondo plano (pluviómetro) ubicado en una estación meteorológica y midiendo en horarios fijos la altura del agua recogida en milímetros. Cada milímetro de altura en el pluviómetro equivale a 1 L de agua por m².

````python
# Desarrolle su ejercicio aquí

def ingresar_datos():
    datos = []

    while True:
        # Solicitar el año
        año = int(input("Ingresa el año: "))

        # Crear un diccionario para almacenar los datos de un año
        año_data = {"AÑO": año}

        # Solicitar las precipitaciones de cada mes
        for mes in ["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO", "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"]:
            precip = float(input(f"Ingresa la precipitación en mm para {mes}: "))
            año_data[mes] = precip

        # Agregar el año y sus datos a la lista de datos
        datos.append(año_data)

        # Preguntar si se desea ingresar otro año
        continuar = input("¿Quieres ingresar otro año? (si/no): ").strip().lower()
        if continuar != 'si':
            break

    return datos

# Función principal para realizar los cálculos
def realizar_calculos(datos):
    # 1. Promedio anual de precipitación para cada año
    promedio_anual = []
    for año_data in datos:
        total_anual = sum(año_data[mes] for mes in año_data if mes != "AÑO")
        promedio_anual.append({"AÑO": año_data["AÑO"], "Promedio Anual": total_anual / 12})

    # 2. Promedio mensual de precipitación para cada mes
    meses = ["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO", "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"]
    promedio_mensual = {mes: sum(año_data[mes] for año_data in datos) / len(datos) for mes in meses}

    # 3. El mes y año con menor precipitación
    min_precipitacion = float('inf')
    min_mes = ""
    min_año = ""
    for año_data in datos:
        for mes in meses:
            if año_data[mes] < min_precipitacion:
                min_precipitacion = año_data[mes]
                min_mes = mes
                min_año = año_data["AÑO"]

    # 4. El año más lluvioso
    año_mas_lluvioso = max(datos, key=lambda año_data: sum(año_data[mes] for mes in meses if mes != "AÑO"))
    total_precipitacion_año_mas_lluvioso = sum(año_mas_lluvioso[mes] for mes in meses if mes != "AÑO")

    # 5. El mes más lluvioso
    mes_mas_lluvioso = max(promedio_mensual, key=promedio_mensual.get)
    precipitacion_mes_mas_lluvioso = promedio_mensual[mes_mas_lluvioso]

    # Resultados
    print("\nPromedio anual de precipitación para cada año:")
    for item in promedio_anual:
        print(f"Año: {item['AÑO']}, Promedio Anual: {item['Promedio Anual']:.2f} mm")

    print("\nPromedio mensual de precipitación:")
    for mes, promedio in promedio_mensual.items():
        print(f"{mes}: {promedio:.2f} mm")

    print(f"\nEl mes y año con menor precipitación:")
    print(f"Año: {min_año}, Mes: {min_mes}, Precipitación: {min_precipitacion} mm")

    print(f"\nEl año más lluvioso:")
    print(f"Año: {año_mas_lluvioso['AÑO']}, Total Precipitación: {total_precipitacion_año_mas_lluvioso:.2f} mm")

    print(f"\nEl mes más lluvioso:")
    print(f"Mes: {mes_mas_lluvioso}, Precipitación promedio: {precipitacion_mes_mas_lluvioso:.2f} mm")

# Función para ejecutar el programa
def main():
    while True:
        datos = ingresar_datos()
        realizar_calculos(datos)

        # Preguntar al usuario si quiere volver a ejecutar el programa
        continuar = input("¿Quieres volver a ingresar datos y realizar los cálculos? (si/no): ").strip().lower()
        if continuar != 'si':
            print("Gracias por usar el programa. ¡Hasta luego!")
            break

# Ejecutar el programa
main()

````

# (25%) Ejercicio 4

Al final del bloque completaremos un programa que permita calcular la calificación final de un curso de acuerdo al porcentaje de ponderación y a la calificación obtenida en cada tarea.

En este ejercicio debes solicitar al usuario, con la función ```input()```, la cantidad de tareas que tiene la asignatura, luego solicita las ponderaciones de cada tarea, así como la calificación obtenida y el nombre de cada tarea.

Analiza en qué tipo de arreglo es mas conveniente almacenar la información para calcular la calificación global del curso y presentar un resumen del desempeño del usuario. Recuerda usar bucles para hacer más eficiente el algoritmo.

````python
# Desarrolle su ejercicio aquí

# Función para calcular la calificación final del curso
def calcular_calificacion():
    # Solicitar la cantidad de tareas al usuario
    num_tareas = int(input("Ingresa la cantidad de tareas: "))

    # Inicializar listas para almacenar la información
    tareas = []  # Lista que almacenará tuplas con (nombre, ponderación, calificación)

    # Solicitar la información de cada tarea
    for i in range(num_tareas):
        print(f"\nTarea {i+1}:")
        nombre = input("Ingrese el nombre de la tarea: ")
        ponderacion = float(input(f"Ingrese la ponderación de la tarea {nombre} (en porcentaje): "))
        calificacion = float(input(f"Ingrese la calificación obtenida en la tarea {nombre}: "))

        # Almacenar la información de la tarea como una tupla en la lista
        tareas.append((nombre, ponderacion, calificacion))

    # Calcular la calificación final ponderada
    calificacion_final = 0
    for tarea in tareas:
        nombre, ponderacion, calificacion = tarea
        calificacion_final += (calificacion * ponderacion) / 100

    # Mostrar el resumen del desempeño
    print("\nResumen del desempeño del estudiante:")
    for tarea in tareas:
        nombre, ponderacion, calificacion = tarea
        print(f"Tarea: {nombre}, Calificación: {calificacion}, Ponderación: {ponderacion}%")

    # Mostrar la calificación final
    print(f"\nLa calificación final del curso es: {calificacion_final:.2f}")

# Llamar a la función para ejecutar el programa
calcular_calificacion()


# https://github.com/Diego8223/tarea-3-fundamentos-de-programaci-n


````
